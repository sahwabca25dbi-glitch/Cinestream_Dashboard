## Data Quality Issues

1. Duplicate rows found.
2. Missing Director values.
3. Missing AddedDate values.
4. Trailing spaces in Genre.
5. Trailing spaces in Language.
6. Case mismatch in Type.
7. Invalid IMDbScore values.
8. Invalid CriticRating values.
9. Negative SubscribersGainedThousands.
10. Runtime outlier detected.


```python
text_cols = ["Type","Genre","Language","Country","Director","AgeRating","TrendingStatus"]

for col in text_cols:
    df[col] = df[col].astype(str).str.strip()
```

```python
df["Type"] = df["Type"].str.title()
```

```python
df = df.drop_duplicates()
```

```python
df["AddedDate"] = pd.to_datetime(
    df["AddedDate"],
    errors="coerce"
)
```

```python
df.loc[
    (df["IMDbScore"] < 1) |
    (df["IMDbScore"] > 10),
    "IMDbScore"
] = np.nan
```

```python
df.loc[(df["CriticRating"] < 1) | (df["CriticRating"] > 10), "CriticRating"] = np.nan
```

```python
df.loc[df["RuntimeMinutes"] > 1000, "RuntimeMinutes"] = np.nan
```

```python
df.loc[
    df["SubscribersGainedThousands"] < 0,
    "SubscribersGainedThousands"
] = 0
```

```python
df["Profit_Cr"] = (
    df["RevenueCr"]
    - df["ProductionCostCr"]
)
```

```python
df["ROI_Pct"] = (
    df["Profit_Cr"]
    / df["ProductionCostCr"]
) * 100
```

```python
df["Performance_Band"] = np.where(
    df["Profit_Cr"] > 20,
    "Hit",
    np.where(
        df["Profit_Cr"] >= 0,
        "Break-even",
        "Flop"
    )
)
```
